```python
color = ['b', 'g', 'r']
```

```python
for i, col in enumerate(color):
    histogram2 = cv2.calcHist([image], [i], None, [260], [0, 256])
    plt.plot(histogram2, color=col)
    plt.xlim([0, 256])
    
plt.show()
```
